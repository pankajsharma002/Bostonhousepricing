```python
## Independent and Dependent Features

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
##Train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
##Model training
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept

print(regression.coef_)
```

```python
print(regression.intercept_) ## what is it
```

```python
## on which parametres the model has been trained
regression.get_params()
```

```python
## prediction with Test Data
reg_pred = regression.predict(X_test)
reg_pred
```

```python
## plot a scatter plot for the prediction

plt.scatter(y_test,reg_pred)
```

```python
residual = y_test-reg_pred
residual
```

```python
sns.displot(residual,kind="kde")
```

```python
## scatter plot wrt prediction and residual

plt.scatter(reg_pred,residual)

# data is distributed uniformly
```

```python
# performance metrics(mse,mae,rmse)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
      
```

## R square and adjusted R square

formula R^2 = 1- (sum of square residual / sum of square total)

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

## Adjusted R2 = 1- [(1-R2)*(n-1)/(n-k-1)]

where :
R2: The R2 of the model, n: The number of observations, k: the number of predictor variables

```python
1-(1-score)*(len(y_test)-1)/ (len(y_test)-X_test.shape[1]-1)
```
